```python
# Question 1

%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

line1 = 0
plt.subplots_adjust(left=0.07, bottom=0.25,right=0.98)
init_tau1=0
init_tau2=0


f = plt.imread("cameraman.tif","gray")
plt.subplot(131)
plt.title('Input image')
plt.imshow(f,cmap='gray')

# Input output characteristics
y = np.arange(0,256,1)
x = np.linspace(start=0, stop=255,num=256)

def plot_ioc(tau1,tau2):
    y = np.arange(0,256,1)
    y[y<tau1]=tau1
    y[y>tau2]=tau2
    global line1
    plt.subplot(133)
    plt.title('Input Output characteristics')
#     plt.xlabel('X')
#     plt.ylabel('Y')
    line1, = plt.plot(x, y, 'r-')
    
def plot_image(tau1,tau2):
    g=f.copy()
    g[f>tau2]=tau2
    g[f<tau1]=tau1
    plt.subplot(132)
    plt.title('Output image')
    plt.imshow(g,cmap='gray',vmax=255,vmin=0)

plot_image(init_tau1,init_tau2)
plot_ioc(init_tau1,init_tau2)
# Left, Bottom, Height, Length
axt1 = plt.axes([0.20, 0.13, 0.65, 0.03], facecolor='lightgoldenrodyellow')
axt2 = plt.axes([0.20, 0.08, 0.65, 0.03], facecolor='lightgoldenrodyellow')

slider1 = Slider(axt1,'Slider for tau1',valmax=255,valmin=0, valinit=init_tau1, valstep=1)
slider2 = Slider(axt2,'Slider for tau2',valmax=255,valmin=0, valinit=init_tau2, valstep=1, slidermin=slider1)
slider1.slidermax=slider2

def update(val):
    a=int(slider1.val)
    b=int(slider2.val)
    if a!=0:
         tau1=a 
    else:tau1=init_tau1
    if b!=0: 
        tau2=b 
    else:tau2=init_tau2
    line1.remove()
    plot_image(tau1,tau2)
    plot_ioc(tau1,tau2)

slider1.on_changed(update)
slider2.on_changed(update)
plt.show()
```

```python
# Question 5

%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from mpl_toolkits.mplot3d import Axes3D

pi = np.pi

m=np.arange(1,257,1)
n=np.arange(1,257,1)
phi = pi/4

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X,Y = np.meshgrid(m,n)
f = np.cos(2*pi*0.2*X + 2*pi*0.3*Y + phi)
# f = np.cos(2*pi*0.002*X + 2*pi*0.003*Y + phi)

plt.subplots_adjust(left=0.15, bottom=0.3)
plt.title('2D sinusoid')
plt.xlabel('X axis')
plt.ylabel('Y axis')
a = ax.plot_surface(X,Y,f,cmap='gray')

axis = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor='lightgoldenrodyellow')
slider = Slider(axis,'Slider for phi',valmax=2*pi,valmin=0, valinit=phi, valstep=0.1)
slider.closedmax = False
def update(val):
    phi=slider.val
    global a
    a.remove()
    f = np.cos(2*pi*0.2*X + 2*pi*0.3*Y + phi)
#     f = np.cos(2*pi*0.002*X + 2*pi*0.003*Y + phi)
    a = ax.plot_surface(X,Y,f,rcount=150,ccount=150,cmap='gray')
    
slider.on_changed(update)

plt.show()
```
