```python
img = mpimg.imread('cameraman.tif')
plt.figure()

imgplot = plt.imshow(img,vmin=0,vmax=255,cmap='gray')
plt.show()
```

```python
img2 = img.astype('float64')


# Normalize the image
out = (img2-np.min(img2))/(np.max(img2)-np.min(img2))
img_org = out.copy();

out = out + 0.3
out[out>1.0]=1.0
out[out<0.0]=0.0
out_img=np.uint8(out*255)
plt.figure()
imgplot = plt.imshow(img_org,cmap='gray')
plt.show()

# write the image to memory
path = 'cameraman_bright.png'
plt.imsave(path,out_img,cmap='gray',vmax=255,vmin=0)
```
